```python
def climbStairs( n):
        prev = 1
        prev2 = 0
        for i in range(1, n+1):
            curi = prev + prev2
            prev2 = prev
            prev = curi
        return prev 
```

```python
climbStairs(7)
```

```python
def climbStairs(n: int) -> int:
        prev = 1
        prev_prev = 0
        for _ in range(n):
            prev_prev, prev = prev, prev + prev_prev
        return prev
```

```python
climbStairs(7)
```
